```python
import pandas as pd
import numpy as np
from datetime import datetime

#test date for insertion
testdate = datetime.strptime("4/14/2014", "%m/%d/%Y")

#path to "UW_Trip_Data_Full_Headers.csv". You'll need to change this.
toFullHeadersPath = "/Users/fineiskid/Desktop/Access_Analysis_Rproject_local/data/UW_Trip_Data_FullHeaders.csv"

#grab lines 700,000 through 800,000 from 18 month data file with headers
data = pd.read_csv(toFullHeadersPath,
                   header = True, skiprows = np.union1d(range(0,700000), range(800001, 3493640)))
headers = ["ServiceDate", "Run", "ProviderId", "EvOrder", "EvId", "ReqTime",
           "SchTime", "ReqLate", "Activity", "ETA", "DwellTime", "StreetNo",
           "OnStreet", "City", "LON", "LAT", "BookingId", "SchedStatus", "SubtypeAbbr","FundingSourceId1",
           "PassOn", "SpaceOn", "PassOff", "SpaceOff", "ClientId", "MobAids"]
data.columns = headers
data.ClientId = np.int64(data.ClientId)
data.ServiceDate = [datetime.strptime(data.ServiceDate[i], "%m/%d/%Y %H:%M:%S") for i in range(0, data.shape[0])]
Apr_14 =data[(data["ServiceDate"] == testdate) & ((data["ProviderId"]==6) | (data["ProviderId"]==5))]
```

```python
#Coordinate boundaries of King County, WA:
upper_right = np.array([49.020430, -116.998768])
lower_left = np.array([45.606961, -124.974842])
minlat = lower_left[0]; maxlat = upper_right[0]
minlon = lower_left[1]; maxlon = upper_right[1]
```

```python
rides = Apr_14.Run.unique()
print len(rides)
if all([type(rides[i])==str for x in range(0,rides.shape[0])]) == False:
    print("WARNING: One of your bus Run ID's isn't a string.")
```

```python
ctr = 0
Apr_14.ClientId[Apr_14.ClientId == -9223372036854775808] = 0 #NA ClientIDs were being treated as -Inf
for ride in rides:
    this_ride = Apr_14[Apr_14["Run"]==ride]
    flag = "OK"
    nrow = this_ride.shape[0]
    lat_coords = this_ride.LAT
    lon_coords = this_ride.LON
    if (any(lat_coords < minlat)) | (any(lon_coords < minlon)) | (any(lat_coords > maxlat)) | (any(lon_coords > maxlon)):
        flag = "BAD_COORDS"
    
    if ~((this_ride.Activity.iloc[0] == 4) & (this_ride.Activity.iloc[-1] == 3)):
        if(flag != "OK"):
            flag = flag + " ROUTE_FINISH_ERROR"
        else:
            flag = "ROUTE_FINISH_ERROR"
    if all(this_ride.LON == this_ride.LON.iloc[0]):
        if(flag != "OK"):
            flag = flag + " NO_MOVEMENT"
        else:
            flag = "NO_MOVEMENT"
    if this_ride.shape[0] == 2:
        if(flag != "OK"):
            flag = flag + " TWO_ROWS"
        else:
            flag = "TWO_ROWS"
    print "Ride %s: flag = %s." %(ride, flag)
    
    if flag == "OK":
        if ctr == 0:
            this_ride.to_csv("../data/single_clean_day.csv", mode ='a', header=True, index = False)
            ctr = 1
        else:
            this_ride.to_csv("../data/single_clean_day.csv", mode='a', header=False, index = False)
                             
        
        
        
```
